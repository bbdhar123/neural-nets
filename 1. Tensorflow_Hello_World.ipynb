### Load tensorflow

```python
import tensorflow as tf
```

```python
tf.__version__
```

### Create a constant tensor

```python
hello = tf.constant('Hello World')
```

```python
print(hello)
```

### Defining shape

```python
hello_2 =  tf.constant('Hello World', shape=[6,2])
print(hello_2)
```

```python
a = tf.constant([1.2])
b = tf.constant([3.4])
c = tf.add(a,b)
print(c)
```

<center><h2>How do I print Hello World</h2></center>

Start a session

```python
sess = tf.Session()
```

```python
print(sess.run(hello))
```

```python
print(sess.run(hello_2))
```

```python
sess.run(c)
```

Close the session

```python
sess.close()
```
